# Value converstion

```python
      
def covert_value(value):
    if "M" in value:
        new_value = value.replace("€","").replace("M","")
        new_value = float(new_value)*1000000
    elif "K" in value:
        new_value = value.replace("€","").replace("K","")
        new_value = float(new_value)*1000
    else:
        new_value = value.replace("€","")
        new_value = float(new_value)
        
    return new_value

cat_df1['Value'] = cat_df1['Value'].apply(covert_value)
```

```python
#cleaning data in 'LS'...
def convert_position(nb):
    if isinstance(nb, int):
        return nb 
    
    if '-' in nb:
        score, bonus = map(int, nb.replace('-', '').split("+"))
        total = score - bonus
    else:
        score, bonus = map(int, nb.split("+"))
        total = score + bonus
    
    return total

for i in value_num:
    cat_df1[value_num]  = df1[value_num].applymap(convert_position)
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
print("R2_score of training data: ",r2_score(y_train, predictions))
```

```python
predictions_test = lm.predict(X_test)
print("R2_score of testing data: ",r2_score(y_test, predictions_test))
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
print("root mean squared Error: ",rmse)
```
